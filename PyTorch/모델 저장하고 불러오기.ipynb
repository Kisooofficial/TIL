```python
import torch
import torchvision.models as models

model = models.vgg16(weights = 'IMAGENET1K_V1')
torch.save(model.state_dict(), 'model_weights.pth') ## state_dict() 내부 상태 사전에 저장함. 이 상태값들은 torch.save 메소드를 통해 저장
```

```python
model = models.vgg16()
model.load_state_dict(torch.load('model_weights.pth')) ## 동일한 형태의 모델을 먼저 만들고 매개변수 불러옴
model.eval()
```

```python
torch.save(model, 'model.pth')
```
